# **Implementation of Synthetic Minor Oversampling Technique(SMOTE) to solve Class imbalance**

```python
from imblearn.over_sampling import SMOTE
from collections import Counter
```

```python
#Fitting SMOTE to the training set
smote = SMOTE(random_state = 27)
smote_X_train, smote_Y_train = smote.fit_resample(X_train,Y_train)

print("Before sampling, Class Distribution:- ",Counter(Y_train))
print("After sampling, Class Distribution:- ",Counter(smote_Y_train))
```

```python
class_distribution = Counter(smote_Y_train)

# Create a bar chart
plt.bar(class_distribution.keys(), class_distribution.values())
plt.title('Class Distribution After SMOTE')
plt.xlabel('Class')
plt.ylabel('Count')

# Add labels to the bars
for class_label, count in class_distribution.items():
    plt.text(class_label, count, str(count), ha='center', va='bottom')

# Show the plot
plt.show()
```

```python
#Importing Dependencies
from sklearn.tree import DecisionTreeClassifier, plot_tree 
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report, roc_curve, auc
import numpy as np
import pickle
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
import warnings

#Model Training
# Create a Decision Tree model
model_DT = DecisionTreeClassifier(random_state=2)

# Train the model
model_DT.fit(smote_X_train, smote_Y_train)


#Model Evaluation 
# Accuracy on training data
X_train_prediction = model_DT.predict(smote_X_train)
training_data_accuracy = accuracy_score(X_train_prediction, smote_Y_train)
print('Accuracy on Training data :', training_data_accuracy)

# Accuracy on test data
X_test_prediction = model_DT.predict(X_test)
test_data_accuracy = accuracy_score(X_test_prediction, Y_test)
print('Accuracy on Test data :', test_data_accuracy)

# Confusion matrix and classification report
print("Confusion Matrix:\n", confusion_matrix(X_test_prediction, Y_test))
print("\nClassification Report:\n", classification_report(X_test_prediction, Y_test))
#ROC Curve & AUC
# Predicted probabilities
Y_train_prob = model_DT.predict_proba(smote_X_train)[:, 1]
Y_test_prob = model_DT.predict_proba(X_test)[:, 1]

# ROC Curve and AUC
fpr_train, tpr_train, _ = roc_curve(smote_Y_train, Y_train_prob)
fpr_test, tpr_test, _ = roc_curve(Y_test, Y_test_prob)

roc_auc_train = auc(fpr_train, tpr_train)
roc_auc_test = auc(fpr_test, tpr_test)

# Plot
plt.figure(figsize=(8, 6))
plt.plot(fpr_train, tpr_train, color='darkorange', lw=2, label=f'Training ROC curve (AUC = {roc_auc_train:.2f})')
plt.plot(fpr_test, tpr_test, color='cornflowerblue', lw=2, label=f'Test ROC curve (AUC = {roc_auc_test:.2f})')
plt.plot([0, 1], [0, 1], color='gray', lw=2, linestyle='--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating Characteristic (ROC) Curve')
plt.legend(loc='lower right')
plt.show()

#Making a Predictive System
# Initialize the scaler
scaler = StandardScaler()

# Fit the scaler with your training data (add this before model training)
scaler.fit(smote_X_train)

# Sample input for prediction
input_data = (0, 77, 0, 0, 4, 31.7, 6.5, 280)

# Convert and reshape
input_data_as_numpy_array = np.asarray(input_data).reshape(1, -1)

# Standardize using the same scaler
with warnings.catch_warnings():
    warnings.simplefilter("ignore")
    standardized_input = scaler.transform(input_data_as_numpy_array)

# Standardize using the same scaler
standardized_input = scaler.transform(input_data_as_numpy_array)

# Make prediction
prediction = model_DT.predict(standardized_input)

if prediction[0] == 0:
    print('The Person does not have diabetes')
else:
    print('The Person has diabetes')

##
# Plot the Decision Tree
plt.figure(figsize=(20,10))
plot_tree(model_DT, filled=True, class_names=['No Diabetes', 'Diabetes'], rounded=True)
plt.title("Decision Tree Visualization")
plt.show()



#Saving the trained model
# Save the trained model and scaler
file_name = 'trained_model_using_DT.sav'
pickle.dump(model_DT, open(file_name, 'wb'))

# Save the scaler
standard_data = 'standard_scaler.sav'
pickle.dump(scaler, open(standard_data, 'wb'))

# Load the model and scaler (if needed later)
loaded_model_DT = pickle.load(open('trained_model_using_DT.sav', 'rb'))
loaded_scaler = pickle.load(open('standard_scaler.sav', 'rb'))
```
